```python
import numpy as np
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt

%matplotlib inline
# note that this can cause problems with saving figure to file - 
# savefig() must be called from the same cell as creation method
```

```python
np.random.seed(15)

# create a list of 10 digits evenly distributed between 1 and 40
x = np.linspace(1, 40, 10)

y1 = x / (np.random.rand(10)*2)
y2 = x / (np.random.rand(10)*3)

line1 = Line2D(x, y1, label='Son of a Line')
line2 = Line2D(x, y2, label='Another Line')

# Instantiate a figure and axes
fig, ax = plt.subplots()

ax.add_line(line1)
ax.add_line(line2)

# by default x and y axis values range between 0 and 1
ax.autoscale()
```

You might be stumbled, why we don't see any lines. It is because, the x and y boundaries of this subplot axes ranges between 0 and 1, meanwhile the numbers we generated do not fit into the scale. Hence we may call a helper function. Autoscale to fix it

Now let's see if we can refer to all or one of the lines on this cart
To do that, we need to call the container that holds reference to Line2D artists


```python
ax.lines
```

As you can see, lines container returns two Line2D objects. Because we have two lines on the charting area, or axes as it is called. Since we know how to call lines container, we can address them easily. So, let's do it

```python
# We will get a little ahead of us and add a color to the line
# But don't worry, we have a special class where we will visit
# line properties
ax.lines[0].set_color('r')

ax.figure
```

We can do a lot of things with Line2D objects, but more on that in our next class. If you have any questions, do let me know and bye for now.

```python
fig.set_size_inches(15,5)
fig
```

```python
# Let's create handlers for the two lines
# in the axess, as we will be referring to them
# soon
l1 = ax.lines[0]
l2 = ax.lines[1]

# how about we change line1 opacity?
l1.set_alpha(0.3)

fig
```

```python
for line in ax.lines:
    # make all the lines red
    line.set_color('red')    
    # make all the lines wider
    line.set_linewidth(2)
    
ax.figure
```

```python
l1.set_linestyle('dashed')
fig
```

```python
l2.set_marker('*')
l2.set_markersize(20)
l2.set_markeredgecolor('red')
l2.set_markerfacecolor('white')
l2.set_markeredgewidth(1)
l2.set_markevery(1)
l2.set_sketch_params(0.9)

fig
```

```python
l1.set_clip_on(False)
# here we are saying, turn off the clipping from the dashed line
ax.set_xlim(0, 39)
# set xaxis boundary to 39
fig
```

```python
ax.set_title("Some line chart")
ax.set_xlabel('This is x axis')
ax.set_ylabel('This is y axis')

ax.figure
```
