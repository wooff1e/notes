```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100) # 100 values evenly spaced in [0, 10] 
y = 4 + 2 * np.sin(2 * x)
fig = plt.figure()
'''
additional parameters:
    figsize	(width,height) # in inches
    dpi, facecolor, edgecolor, linewidth
'''
# Axes object is the region of the image with the data space
ax=fig.add_axes([0,0,1,1])  # [left, bottom, width, height] - all quantities are in fractions of figure width and height
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
# The Axes contains two (or three in the case of 3D) Axis objects. 
ax.plot(x,y)

plt.show()
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])

line1 = ax.plot(x,y)
line2 = ax.plot(x+1,y,'go--')   # # dash line with green colour and circle marker
ax.legend(labels=['first', 'second'], loc=1)

plt.show()
```

```python
# multiple plots
'''
The subplot() function returns the axes object at a given grid position:
plt.subplot(subplot(nrows, ncols, index)

The indexes can also be given as single, concatenated, threedigitnumber:
subplot(2, 3, 3) --> subplot(233) 

Creating a subplot will delete any pre-existing subplot that overlaps with it beyond sharing a boundary!
'''
# plot a line, implicitly creating a subplot(111)
plt.plot([1,2,3])
# now create a subplot which will overlap the first plot, that plot (and its axes), will be removed
plt.subplot(211)
plt.plot(range(12))
plt.subplot(212, facecolor='y')
plt.plot(range(12))
```

```python
# The add_subplot() function of the figure class will not overwrite the existing plot:
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot([1,2,3])
ax2 = fig.add_subplot(221, facecolor='y')
ax2.plot([1,2,3])
```

```python
# You can add an insert plot in the same figure by adding another axes object in the same figure canvas.
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x,np.cos(x),'r')
axes1.set_title('sine')
axes2.set_title("cosine")
plt.show()
```

```python
# grid() method
fig, axes = plt.subplots(1,3, figsize = (12,4))
x = np.arange(1,11)
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls = '-.', lw = 0.25)
axes[1].set_title('custom grid')
axes[2].plot(x,x)
axes[2].set_title('no grid')
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
x = np.arange(1,5)
axes[0].plot( x, np.exp(x))
axes[0].plot(x,x**2)
axes[0].set_title("Normal scale")
axes[1].plot (x, np.exp(x))
axes[1].plot(x, x**2)
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")
axes[0].xaxis.labelpad = 10
axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.plot([1,2,3,4,5])
plt.show()
```

```python
# set limits
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])

x = np.arange(1,10)
a1.plot(x, np.exp(x),'r')
a1.set_title('exp')
a1.set_ylim(0,10000)
a1.set_xlim(0,10)
plt.show()
```

```python
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
y = np.sin(x)
ax.plot(x, y)
ax.set_xlabel('angle')
ax.set_title('sine')
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(['zero','two','four','six'])
ax.set_yticks([-1,0,1])
plt.show()
```

```python
# Twin axes
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```
