```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
def experiment(nr_repeats):
    nr_heads = []
    fig, ax = plt.subplots()

    for i in range(nr_repeats):
        throw = random.randint(2, size=(5))
        #print(throw)
        heads = np.sum(throw)
        nr_heads.append(heads)

    x = [
        np.sum(np.array(nr_heads)==0), 
        np.sum(np.array(nr_heads)==1), 
        np.sum(np.array(nr_heads)==2), 
        np.sum(np.array(nr_heads)==3), 
        np.sum(np.array(nr_heads)==4), 
        np.sum(np.array(nr_heads)==5),
    ]

    ind = np.arange(6)
    ax.bar(ind, x, 1)
    plt.show()
```

```python
experiment(1)
```

```python
experiment(10)
```

```python
experiment(100)
```

```python
experiment(1000)
```

```python
experiment(10000)
```

```python
experiment(100000)
```
